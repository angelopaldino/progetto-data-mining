```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset completo con metriche Train e Test per i modelli RF e Balanced RF
rf_data = [
    # Random Forest
    {"name": "Random Forest", "set": "Train", "accuracy": 1.0000, "precision": 1.0000, "recall": 1.0000, "f1": 1.0000},
    {"name": "Random Forest", "set": "Test",  "accuracy": 0.6212, "precision": 0.6538, "recall": 0.5514, "f1": 0.5773},
    
    # Random Forest GridSearch
    {"name": "RF GridSearch", "set": "Train", "accuracy": 0.9903, "precision": 0.9826, "recall": 0.9912, "f1": 0.9867},
    {"name": "RF GridSearch", "set": "Test",  "accuracy": 0.6578, "precision": 0.6368, "recall": 0.5872, "f1": 0.6013},
    
    # Balanced Random Forest
    {"name": "Balanced RF", "set": "Train", "accuracy": 0.8474, "precision": 0.8291, "recall": 0.8707, "f1": 0.8447},
    {"name": "Balanced RF", "set": "Test",  "accuracy": 0.6010, "precision": 0.5481, "recall": 0.5593, "f1": 0.5438},
]

df_rf = pd.DataFrame(rf_data)

# Plot per ogni metrica
metrics = ['accuracy', 'precision', 'recall', 'f1']

for metric in metrics:
    plt.figure(figsize=(10, 6))
    for set_type in ['Train', 'Test']:
        subset = df_rf[df_rf['set'] == set_type]
        plt.bar(
            [x + (0.2 if set_type == "Test" else -0.2) for x in range(len(subset))],
            subset[metric],
            width=0.4,
            label=set_type
        )

    plt.xticks(range(len(subset)), subset['name'], rotation=15)
    plt.title(f'Random Forest Models - {metric.capitalize()} (Train vs Test)')
    plt.ylabel(metric.capitalize())
    plt.ylim(0, 1.05)
    plt.legend()
    plt.grid(axis='y')
    plt.tight_layout()
    plt.savefig(f"../results/classification_category/confronto/randomforest_{metric}.png")
    plt.show()

```
