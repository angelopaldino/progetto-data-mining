# Classificazione multiclasse analisi
In questa sezione viene analizzato il comportamento dei modelli applicati al dataset originale quindi con 18 classi e dei modelli applicato al dataset aggregato(aggregando opportunamente le classi).

# Confronto decision tree sia nel caso di criterio gini che entropy

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import e struttura dati per tutte le combinazioni del Decision Tree (gini/entropy) su dataset aggregato/non aggregato

data = [
    # Non aggregato
    {"model": "Decision Tree (Entropy)", "agg": "No", "set": "Train", "accuracy": 1.0, "precision": 1.0, "recall": 1.0, "f1": 1.0},
    {"model": "Decision Tree (Entropy)", "agg": "No", "set": "Test", "accuracy": 0.4823, "precision": 0.4720, "recall": 0.4270, "f1": 0.4389},
    {"model": "Decision Tree (Gini)", "agg": "No", "set": "Train", "accuracy": 1.0, "precision": 1.0, "recall": 1.0, "f1": 1.0},
    {"model": "Decision Tree (Gini)", "agg": "No", "set": "Test", "accuracy": 0.4823, "precision": 0.4720, "recall": 0.4270, "f1": 0.4389},

    # Aggregato
    {"model": "Decision Tree (Entropy)", "agg": "S", "set": "Train", "accuracy": 1.0, "precision": 1.0, "recall": 1.0, "f1": 1.0},
    {"model": "Decision Tree (Entropy)", "agg": "S", "set": "Test", "accuracy": 0.4571, "precision": 0.4336, "recall": 0.4121, "f1": 0.4207},
    {"model": "Decision Tree (Gini)", "agg": "S", "set": "Train", "accuracy": 1.0, "precision": 1.0, "recall": 1.0, "f1": 1.0},
    {"model": "Decision Tree (Gini)", "agg": "S", "set": "Test", "accuracy": 0.5290, "precision": 0.5169, "recall": 0.4828, "f1": 0.4956},
]

df = pd.DataFrame(data)

df_melted = df.melt(id_vars=["model", "agg", "set"], var_name="metric", value_name="score")

g = sns.catplot(
    data=df_melted,
    x="model",
    y="score",
    hue="set",
    col="metric",
    row="agg",
    kind="bar",
    height=4,
    aspect=1.3,
    palette="Set2"
)

g.set_titles(row_template="{row_name} Aggregato", col_template="{col_name}")
g.set_axis_labels("Modello", "Valore metrica")
g.set_xticklabels(rotation=15)
g.tight_layout()
plt.subplots_adjust(top=0.9)
g.fig.suptitle("Confronto Decision Tree (Gini vs Entropy) su dataset Aggregato e Non")
plt.savefig("../results/classification_category/confronto_originale_aggregato/decisiontree.png")
plt.show()

```
