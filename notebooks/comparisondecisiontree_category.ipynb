```python
import pandas as pd
import matplotlib.pyplot as plt

# Creazione DataFrame con tutte le metriche (Train e Test) dei modelli multiclass
multiclass_data = [
    # Default Entropy
    {"name": "DT Entropy Default", "set": "Train", "accuracy": 1.0000, "precision": 1.0000, "recall": 1.0000, "f1": 1.0000},
    {"name": "DT Entropy Default", "set": "Test",  "accuracy": 0.4571, "precision": 0.4336, "recall": 0.4121, "f1": 0.4207},
    
    # Default Gini
    {"name": "DT Gini Default", "set": "Train", "accuracy": 1.0000, "precision": 1.0000, "recall": 1.0000, "f1": 1.0000},
    {"name": "DT Gini Default", "set": "Test",  "accuracy": 0.5290, "precision": 0.5169, "recall": 0.4828, "f1": 0.4956},
    
    # GridSearch Gini
    {"name": "DT Gini GridSearch", "set": "Train", "accuracy": 0.6423, "precision": 0.7353, "recall": 0.5857, "f1": 0.6234},
    {"name": "DT Gini GridSearch", "set": "Test",  "accuracy": 0.5265, "precision": 0.6188, "recall": 0.4467, "f1": 0.4752},
]

df_multi = pd.DataFrame(multiclass_data)

# Plot per ogni metrica: confronto Train vs Test
metrics = ['accuracy', 'precision', 'recall', 'f1']

for metric in metrics:
    plt.figure(figsize=(10, 6))
    for set_type in ['Train', 'Test']:
        subset = df_multi[df_multi['set'] == set_type]
        plt.bar(
            [x + (0.2 if set_type == "Test" else -0.2) for x in range(len(subset))],
            subset[metric],
            width=0.4,
            label=set_type
        )

    plt.xticks(range(len(subset)), subset['name'], rotation=15)
    plt.title(f'Multiclass Decision Tree - {metric.capitalize()} (Train vs Test)')
    plt.ylabel(metric.capitalize())
    plt.ylim(0, 1.05)
    plt.legend()
    plt.grid(axis='y')
    plt.tight_layout()
    plt.savefig(f"../results/classification_category/comparison/decisiontree_{metric}_train_test.png")
    plt.show()

```
