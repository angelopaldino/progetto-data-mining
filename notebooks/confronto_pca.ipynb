# Classificazione multiclasse

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset con metriche dopo PCA
pca_models_data = [
    {
        "name": "MLP (post-PCA)",
        "set": "Train", "accuracy": 0.9199, "precision": 0.9202, "recall": 0.9013, "f1": 0.9100,
    },
    {
        "name": "MLP (post-PCA)",
        "set": "Test", "accuracy": 0.6881, "precision": 0.6959, "recall": 0.6205, "f1": 0.6424,
    },
    {
        "name": "SVM (post-PCA)",
        "set": "Train", "accuracy": 0.9394, "precision": 0.9452, "recall": 0.9135, "f1": 0.9275,
    },
    {
        "name": "SVM (post-PCA)",
        "set": "Test", "accuracy": 0.6843, "precision": 0.7757, "recall": 0.6235, "f1": 0.6660,
    },
    {
        "name": "Random Forest (post-PCA)",
        "set": "Train", "accuracy": 1.0000, "precision": 1.0000, "recall": 1.0000, "f1": 1.0000,
    },
    {
        "name": "Random Forest (post-PCA)",
        "set": "Test", "accuracy": 0.6919, "precision": 0.7387, "recall": 0.6344, "f1": 0.6632,
    },
]

df_pca_models = pd.DataFrame(pca_models_data)

# Plot metriche
metrics = ['accuracy', 'precision', 'recall', 'f1']

for metric in metrics:
    plt.figure(figsize=(10, 6))
    for set_type in ['Train', 'Test']:
        subset = df_pca_models[df_pca_models['set'] == set_type]
        plt.bar(
            [x + (0.2 if set_type == "Test" else -0.2) for x in range(len(subset))],
            subset[metric],
            width=0.4,
            label=set_type
        )

    plt.xticks(range(len(subset)), subset['name'], rotation=20, ha='right')
    plt.title(f'Modelli post-PCA - {metric.capitalize()} (Train vs Test)')
    plt.ylabel(metric.capitalize())
    plt.ylim(0, 1.05)
    plt.legend()
    plt.grid(axis='y')
    plt.tight_layout()
    plt.savefig(f"../results/classification_category/confronto/pca_{metric}.png")
    plt.show()
```
