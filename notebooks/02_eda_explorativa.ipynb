```python
# 游릱 Cell 1: Importa librerie
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Per una migliore leggibilit dei grafici
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("pastel")
```

```python
# 游릱 Cell 2: Caricamento dati
df = pd.read_csv("../data/processed/billionaires_clean.csv") 
df.head()
```

```python
# 游릱 Cell 3: Info generali
print("游릭 Shape:", df.shape)
print("游릭 Columns:\n", df.columns.tolist())
df.info()
```

```python
# 游릱 Cell 4: Statistiche descrittive (numeriche)
df.describe().T
```

```python
# 游릱 Cell 5: Istogrammi per variabili numeriche
numeric_cols = ['age','finalWorth']
df[numeric_cols].hist(bins=30, figsize=(12, 6))
plt.tight_layout()
plt.show()
```

```python
# 游릱 Cell 6: Analisi variabili categoriche
categorical_cols = ['country', 'industries', 'gender']

for col in categorical_cols:
    print(f"\n游늵 Top categorie per {col}:")
    print(df[col].value_counts().head(10))

    plt.figure(figsize=(8,4))
    sns.countplot(data=df, y=col, order=df[col].value_counts().head(10).index)
    plt.title(f"Top {col}")
    plt.show()
```

```python
# 游릱 Cell 7: Scatterplot Et vs Patrimonio
plt.figure(figsize=(8,6))
sns.scatterplot(x="age", y="finalWorth", data=df)
plt.title("Et vs Patrimonio")
plt.xlabel("Et")
plt.ylabel("finalWorth (in miliardi)")
plt.show()
```

```python
# 游릱 Cell 8: Boxplot finalWorth per Industry (top 8)
top_industries = df['industries'].value_counts().head(8).index
df_top = df[df['industries'].isin(top_industries)]

plt.figure(figsize=(12,6))
sns.boxplot(data=df_top, x="industries", y="finalWorth")
plt.xticks(rotation=45)
plt.title("Distribuzione del finalWorth per Industries (Top 8)")
plt.show()
```

```python
# 游릱 Cell 9: Heatmap delle correlazioni numeriche
# Questa heatmap 칟 la matrice di correlazione fra tutte le variabili numeriche del dataset: ogni cella ti mostra il coefficiente di Pearson (da -1 a +1) che misura la forza e la direzione (positiva o negativa) dell뗘ssociazione lineare fra due colonne.
corr = df.select_dtypes(include='number').corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matrice di correlazione")
plt.show()
```

```python
# 游릱 Cell 10: Distribuzione geografica
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='barh', figsize=(8,4), color='skyblue')
plt.title("Top 10 Paesi per numero di miliardari")
plt.xlabel("Numero di miliardari")
plt.gca().invert_yaxis()
plt.show()
```

```python
# 游릱 Cell: istogrammi + KDE per le macro났ariabili
#kde: KDE sta per Kernel Density Estimate, ovvero stima della densit tramite kernel. 칃 una tecnica usata per stimare la distribuzione di probabilit continua di una variabile, partendo da un insieme finito di dati.
#cpi_country	Consumer Price Index (CPI) del paese  un indice che misura la variazione media dei prezzi al consumo in un anno, rispetto a un anno base 
#cpi_change_country	Variazione percentuale (year-over-year) del CPI nel paese  indica quanto siano aumentati (o diminuiti) i prezzi al consumo rispetto all뗘nno precedente 
#gdp_country	Gross Domestic Product (GDP) del paese  il valore totale dei beni e servizi prodotti in un anno, spesso espresso in miliardi di USD; misura la dimensione dell뗜conomia nazionale 
#gross_tertiary_education_enrollment	Tasso lordo di iscrizione all뗠struzione terziaria  percentuale di popolazione in et universitaria iscritta a corsi di laurea o formazione post-secondaria 
#gross_primary_education_enrollment_country	Tasso lordo di iscrizione all뗠struzione primaria  percentuale di bambini in et scolare iscritti alle scuole primarie

macro_cols = [
    "cpi_country",
    "cpi_change_country",
    "gdp_country",
    "gross_tertiary_education_enrollment",
    "gross_primary_education_enrollment_country"
]

for col in macro_cols:
    plt.figure(figsize=(8,4))
    # seaborn: istogramma con curva di densit sovrapposta
    sns.histplot(df[col].dropna(), bins=30, kde=True)
    plt.title(f"Distribuzione di `{col}`")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()
```
