```python
import pandas as pd
import matplotlib.pyplot as plt

# Dati completi con metriche Train e Test per i 3 modelli ANN
ann_data = [
    {"name": "ANN base", "set": "Train", "accuracy": 0.7348, "precision": 0.7516, "recall": 0.9215, "f1": 0.8279},
    {"name": "ANN base", "set": "Test",  "accuracy": 0.7042, "precision": 0.7385, "recall": 0.8872, "f1": 0.8060},
    
    {"name": "ANN ReLU (100,)", "set": "Train", "accuracy": 0.7348, "precision": 0.7516, "recall": 0.9215, "f1": 0.8279},
    {"name": "ANN ReLU (100,)", "set": "Test",  "accuracy": 0.7042, "precision": 0.7385, "recall": 0.8872, "f1": 0.8060},
    
    {"name": "ANN GridSearch (100, 50)", "set": "Train", "accuracy": 0.8303, "precision": 0.8521, "recall": 0.9136, "f1": 0.8817},
    {"name": "ANN GridSearch (100, 50)", "set": "Test",  "accuracy": 0.7708, "precision": 0.8142, "recall": 0.8669, "f1": 0.8397},
]

df_ann = pd.DataFrame(ann_data)

# Plot comparativo Train vs Test per ogni metrica
metrics = ['accuracy', 'precision', 'recall', 'f1']

for metric in metrics:
    plt.figure(figsize=(10, 6))
    for set_type in ['Train', 'Test']:
        subset = df_ann[df_ann['set'] == set_type]
        plt.bar(
            [x + (0.2 if set_type == "Test" else -0.2) for x in range(len(subset))],
            subset[metric],
            width=0.4,
            label=set_type
        )

    plt.xticks(range(len(subset)), subset['name'], rotation=15)
    plt.title(f'ANN - {metric.capitalize()} Comparison (Train vs Test)')
    plt.ylabel(metric.capitalize())
    plt.ylim(0, 1)
    plt.legend()
    plt.grid(axis='y')
    plt.tight_layout()
    plt.savefig(f"../results/classification_selfMade/comparison/ann_{metric}.png")
    plt.show()

```
