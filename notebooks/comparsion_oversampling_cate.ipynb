```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset con metriche dopo oversampling sulla classe 2
oversampled_models_data = [
    {
        "name": "MLP (oversampled cls=2)",
        "set": "Train", "accuracy": 0.9197, "precision": 0.9085, "recall": 0.9151, "f1": 0.9108,
    },
    {
        "name": "MLP (oversampled cls=2)",
        "set": "Test", "accuracy": 0.6843, "precision": 0.6526, "recall": 0.6264, "f1": 0.6352,
    },
    {
        "name": "SVM (oversampled cls=2)",
        "set": "Train", "accuracy": 0.9431, "precision": 0.9468, "recall": 0.9262, "f1": 0.9349,
    },
    {
        "name": "SVM (oversampled cls=2)",
        "set": "Test", "accuracy": 0.6843, "precision": 0.7667, "recall": 0.6240, "f1": 0.6647,
    },
    {
        "name": "KNN (oversampled cls=2)",
        "set": "Train", "accuracy": 1.0000, "precision": 1.0000, "recall": 1.0000, "f1": 1.0000,
    },
    {
        "name": "KNN (oversampled cls=2)",
        "set": "Test", "accuracy": 0.6553, "precision": 0.7676, "recall": 0.5981, "f1": 0.6415,
    },
    {
        "name": "Balanced RF (oversampled cls=2)",
        "set": "Train", "accuracy": 0.8044, "precision": 0.7953, "recall": 0.8186, "f1": 0.8008,
    },
    {
        "name": "Balanced RF (oversampled cls=2)",
        "set": "Test", "accuracy": 0.6111, "precision": 0.5577, "recall": 0.5632, "f1": 0.5528,
    },
]

df_oversampled = pd.DataFrame(oversampled_models_data)

# Plot metriche
metrics = ['accuracy', 'precision', 'recall', 'f1']

for metric in metrics:
    plt.figure(figsize=(12, 6))
    for set_type in ['Train', 'Test']:
        subset = df_oversampled[df_oversampled['set'] == set_type]
        plt.bar(
            [x + (0.2 if set_type == "Test" else -0.2) for x in range(len(subset))],
            subset[metric],
            width=0.4,
            label=set_type
        )

    plt.xticks(range(len(subset)), subset['name'], rotation=20, ha='right')
    plt.title(f'Modelli (Oversampling classe 2) - {metric.capitalize()} (Train vs Test)')
    plt.ylabel(metric.capitalize())
    plt.ylim(0, 1.05)
    plt.legend()
    plt.grid(axis='y')
    plt.tight_layout()
    plt.savefig(f"../results/classification_category/comparison/oversampling_{metric}.png")

    plt.show()
```
