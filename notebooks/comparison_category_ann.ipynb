```python

import pandas as pd
import matplotlib.pyplot as plt
# Costruzione del dataset con le metriche ANN multiclass (train vs test)
ann_multi_data = [
    {
        "name": "ANN (64,32) α=0.05",
        "set": "Train", "accuracy": 0.9172, "precision": 0.9088, "recall": 0.9091, "f1": 0.9084,
    },
    {
        "name": "ANN (64,32) α=0.05",
        "set": "Test", "accuracy": 0.6806, "precision": 0.6722, "recall": 0.6167, "f1": 0.6346,
    },
    {
        "name": "ANN (100,50) α=0.01",
        "set": "Train", "accuracy": 0.9172, "precision": 0.9106, "recall": 0.9073, "f1": 0.9085,
    },
    {
        "name": "ANN (100,50) α=0.01",
        "set": "Test", "accuracy": 0.6742, "precision": 0.6739, "recall": 0.6160, "f1": 0.6295,
    },
    {
        "name": "Ensemble MLP (64,32)",
        "set": "Train", "accuracy": 0.9708, "precision": 0.9638, "recall": 0.9739, "f1": 0.9686,
    },
    {
        "name": "Ensemble MLP (64,32)",
        "set": "Test", "accuracy": 0.7020, "precision": 0.6866, "recall": 0.6657, "f1": 0.6694,
    },
]

df_ann_multi = pd.DataFrame(ann_multi_data)

# Plot delle metriche
metrics = ['accuracy', 'precision', 'recall', 'f1']

for metric in metrics:
    plt.figure(figsize=(10, 6))
    for set_type in ['Train', 'Test']:
        subset = df_ann_multi[df_ann_multi['set'] == set_type]
        plt.bar(
            [x + (0.2 if set_type == "Test" else -0.2) for x in range(len(subset))],
            subset[metric],
            width=0.4,
            label=set_type
        )

    plt.xticks(range(len(subset)), subset['name'], rotation=20, ha='right')
    plt.title(f'ANN Multiclass - {metric.capitalize()} (Train vs Test)')
    plt.ylabel(metric.capitalize())
    plt.ylim(0, 1.05)
    plt.legend()
    plt.grid(axis='y')
    plt.tight_layout()
    plt.savefig(f"../results/classification_category/comparison/ann_{metric}.png")
    plt.show()
```
