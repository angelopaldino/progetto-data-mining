# Confronto finale delle metriche ottenute dai modelli di classificazione binaria per la variabile selfMade

```python


# Dati dei modelli
data = {
    "Modello": [
        "Random baseline",
        "ANN (MLPClassifier)",
        "Decision Tree",
        "KNN (k=5)",
        "Naive Bayes",
        "RIPPER (Rule-Based)"
    ],
    "F1 Test": [0.6735, 0.8397, 0.8156, 0.8297, 0.1308, 0.5915],
    "Overfitting Gap (F1)": [0.0195, 0.0420, 0.0086, 0.0407, 0.0281, 0.0094],
    "Accuracy Test": [0.5480, 0.7708, 0.7222, 0.7567, 0.3534, 0.5915],
    "Precision Test": [0.6741, 0.8142, 0.7547, 0.8052, 0.9500, 0.9625],
    "Recall Test": [0.6728, 0.8669, 0.8872, 0.8558, 0.0702, 0.4270]
}
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame(data)
sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 6))

# F1-Score
plt.figure()
sns.barplot(x="F1 Test", y="Modello", hue="Modello", data=df, palette="viridis", legend=False)
plt.title("Confronto F1-Score sui modelli (Test set)")
plt.xlabel("F1-Score")
plt.ylabel("Modello")
plt.tight_layout()
plt.savefig("../results/comparison_selfMade_models/f1_test_comparison.png")

# Overfitting Gap
plt.figure()
sns.barplot(x="Overfitting Gap (F1)", y="Modello", hue="Modello", data=df, palette="magma", legend=False)
plt.title("Overfitting Gap tra Train e Test (F1-Score)")
plt.xlabel("Gap F1 (Train - Test)")
plt.ylabel("Modello")
plt.tight_layout()
plt.savefig("../results/comparison_selfMade_models/overfitting_gap.png")

# Precision e Recall
plt.figure()
df_pr = df[["Modello", "Precision Test", "Recall Test"]].melt(id_vars="Modello", var_name="Metrica", value_name="Valore")
sns.barplot(data=df_pr, x="Valore", y="Modello", hue="Metrica", palette="Set2")
plt.title("Confronto Precision e Recall (Test set)")
plt.xlabel("Valore")
plt.ylabel("Modello")
plt.tight_layout()
plt.legend(title="Metrica")
plt.savefig("../results/comparison_selfMade_models/precision_recall_comparison.png")

print("Grafici salvati in: results/comparison_selfMade_models/")


# Accuracy Test
plt.figure()
sns.barplot(x="Accuracy Test", y="Modello", hue="Modello", data=df, palette="coolwarm", legend=False)
plt.title("Confronto Accuracy sui modelli (Test set)")
plt.xlabel("Accuracy")
plt.ylabel("Modello")
plt.tight_layout()
plt.savefig("../results/comparison_selfMade_models/accuracy_test_comparison.png")
```
